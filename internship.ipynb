```python
import math
import matplotlib.pyplot as plt
def fib(n):
  if(n==1 or n==0):
    return 1
  else:
    return fib(n-1)*n
y_axis=range(0,10)
l=[fib(i) for i in range(0,10)]
x_axis=[math.sin(j) for j in l]
plt.plot(y_axis,x_axis,linestyle='-')
plt.xlabel('sin value')
plt.ylabel('fibonnaci number')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0, 10, 100)
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))
plt.subplot(2,1,1)
plt.plot(x,np.cos(x))
plt.show()
```

```python
from sklearn.datasets import load_iris
from mpl_toolkits import mplot3d
iris=load_iris()
features=iris.data.T
plt.scatter(features[0],features[1],alpha=0.2,s=100*features[3],c=iris.target,cmap='viridis')
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.scatter3D(features[0], features[1], c=iris.target, cmap='viridis')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
species = ('1st pair', '2nd pair', '3rd pair')
sex_counts = {
    'Male': np.array([56,15,55]),
    'Female': np.array([43,66,44]),
}
width = 0.6  # the width of the bars: can also be len(x) sequence


fig, ax = plt.subplots()
bottom = np.zeros(3)

for sex, sex_count in sex_counts.items():
    p = ax.bar(species, sex_count, width, label=sex, bottom=bottom)
    bottom += sex_count

    ax.bar_label(p, label_type='center')

ax.set_title('age of pairs')
ax.legend()

plt.show()
```
